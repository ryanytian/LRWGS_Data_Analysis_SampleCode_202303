# SMALL DELS

```python
small_del_gene_annotated_list_path = \
'/Users/ryanyutian/Desktop/new_panel_TRI_Brain_gene_list/unique_gene_only_annotated/small_dels/'

small_del_gene_list_names = []

os.chdir(small_del_gene_annotated_list_path)
temp_files = sorted([i for i in os.listdir(small_del_gene_annotated_list_path) if 'DS' not in i])

for file_name in temp_files:
    
    temp_df = pd.read_csv(file_name)
    globals()[file_name[:-4]] = temp_df['gene_name'].tolist()
    small_del_gene_list_names.append(file_name[:-4])
```

```python
old_small_del_gene_annotated_list_path = \
'/Users/ryanyutian/Desktop/TRI_Brain_First_Round_Analysis/TRI_Brain_gene_list/unique_gene_only_annotated/small_dels/'

old_small_del_gene_list_names = []

os.chdir(old_small_del_gene_annotated_list_path)
temp_files = sorted([i for i in os.listdir(old_small_del_gene_annotated_list_path) if 'DS' not in i])

for file_name in temp_files:
    
    temp_df = pd.read_csv(file_name)
    globals()['old_' + file_name[:-4] + '_small_del_gene_only_list'] = temp_df['gene_name'].tolist()
    old_small_del_gene_list_names.append('old_' + file_name[:-4] + '_small_del_gene_only_list')
```

```python
shared_gene_list_names = []
old_only_gene_list_names = []
new_gene_list_names = []

percent_differences = {}

for gene_list_name in small_del_gene_list_names:
    
    ###
    shared_list = []
    old_only_list = []
    
    for gene_name in globals()['old_'+ gene_list_name]:
        
        if gene_name in globals()[gene_list_name]:
            
            shared_list.append(gene_name)
            
        else:
            
            old_only_list.append(gene_name)
            
    globals()[gene_list_name[:-15] + '_shared_list'] = shared_list
    globals()[gene_list_name[:-15] + '_old_only_list'] = old_only_list
    
    shared_gene_list_names.append(gene_list_name[:-15] + '_shared_list')
    old_only_gene_list_names.append(gene_list_name[:-15] + '_old_only_list')
    
    perc_diff = 100 * len(old_only_list)/(len(shared_list) + len(old_only_list))
    percent_differences[gene_list_name[:-15]] = perc_diff
                                          
    print('For sample: ' + gene_list_name)
    print('Number of shared genes: ' + str(len(shared_list)))
    print('Number of now filtered out genes: ' + str(len(old_only_list)))
    print('% difference: ' + str(perc_diff) + '%')
    print('\n')
    
    ###
    new_gene_list = []
    
    for gene_name in globals()[gene_list_name]:
        
        if gene_name not in shared_list:
            
            if gene_name not in old_only_list:
                
                new_gene_list.append(gene_name)
            
    print('New genes for sample: ' + gene_list_name)
    print(new_gene_list)  
    print('\n')
    
    globals()[gene_list_name[:-15] + '_new_list'] = new_gene_list
    new_gene_list_names.append(gene_list_name[:-15] + '_new_list')
```
