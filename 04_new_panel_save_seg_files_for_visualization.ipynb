# PANEL OF NORMALS

```python
### Load panel of normals ###

##
panel_path = '/Users/ryanyutian/Desktop/new_panel_TRI_Brain_glioma_sv_processed/panel'

panel_df_names = []

os.chdir(panel_path)
temp_files = sorted([i for i in os.listdir(panel_path) if 'DS' not in i])

for file_name in temp_files:
    
    globals()[file_name[:-4]] = pd.read_csv(file_name)
    panel_df_names.append(file_name[:-4])
```

```python
panel_path = '/Users/ryanyutian/Desktop/new_panel_TRI_BRAIN_seg_files/panel/'

for df_name in panel_df_names:
    
    temp_df = globals()[df_name]
    
    save_df1 = pd.DataFrame({'SAMPLE_ID': ([df_name]*len(temp_df))})
    
    save_df2 = temp_df[['CHROM', 'POS', 'END']].copy()
    save_df2.rename(columns={'POS': 'START'}, inplace=True)
    
    value_df = pd.DataFrame({'VALUE': ([-1]*len(temp_df))})
    
    save_df = pd.concat([save_df1, save_df2, value_df], axis=1)
    
    save_df.to_csv((panel_path + df_name + '.seg'), index=False, sep='\t')
```
